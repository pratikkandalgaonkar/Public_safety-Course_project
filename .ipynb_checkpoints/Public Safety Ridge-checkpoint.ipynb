```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
```

```python
raw_file_path_train = r"C:\Users\prati\Edvancer Data Science\Projects\Public Safety\counterfeit_train.csv"
raw_file_train = pd.read_csv(raw_file_path_train)
raw_file_path_test = r"C:\Users\prati\Edvancer Data Science\Projects\Public Safety\counterfeit_test.csv"
raw_file_test = pd.read_csv(raw_file_path_test)
```

```python
raw_file_train
```

```python
raw_file_test
```

```python
df_1 = raw_file_train.drop(['Medicine_ID'],axis = 1)
df_2 = raw_file_test.drop(['Medicine_ID'],axis = 1)
```

```python
df_1.isnull().sum(), df_2.isnull().sum()
```

```python
df_1_new = df_1.fillna(df_1.median())
df_2_new = df_2.fillna(df_1.median())
```

```python
df_1_new
```

```python
obj_columns =  ['DistArea_ID','Medicine_Type','SidEffect_Level'
                ,'Area_Type','Area_City_Type','Area_dist_level']
dummy_1 = pd.get_dummies(df_1_new, columns = obj_columns, drop_first = True)
dummy_1
```

```python
obj_columns =  ['DistArea_ID','Medicine_Type','SidEffect_Level'
                ,'Area_Type','Area_City_Type','Area_dist_level']
dummy_2 = pd.get_dummies(df_2_new, columns = obj_columns, drop_first = True)
dummy_2
```

```python
dummy_2.nunique()
```

```python
preprocessed_train = dummy_1
```

```python
preprocessed_test = dummy_2
```

```python
target = 'Counterfeit_Sales'
```

```python
x_train = preprocessed_train.drop([target], axis = 1)
y_train = preprocessed_train[target]
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'alpha':np.linspace(0.1,100,100)}
```

```python
model = Ridge(fit_intercept = True)
```

```python
grid = GridSearchCV(model, cv = 10, param_grid = parameters, verbose = 20,scoring = 'neg_mean_squared_error')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.score
```

```python
submission = pd.DataFrame({target:grid.predict(preprocessed_test)})
```

```python
submission.to_csv('my_Submission_Ridge.csv', index = False)
```
