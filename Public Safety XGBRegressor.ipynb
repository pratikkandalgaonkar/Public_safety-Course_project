```python
obj_c = ["DistArea_ID","Medicine_Type","SidEffect_Level",
         "Area_Type","Area_City_Type","Area_dist_level"]
preprocessed_train = pd.get_dummies(train_new, columns = obj_c,drop_first = True)
preprocessed_train
```

```python
obj_c = ["DistArea_ID","Medicine_Type","SidEffect_Level",
         "Area_Type","Area_City_Type","Area_dist_level"]
preprocessed_test = pd.get_dummies(test_new, columns = obj_c,drop_first = True)
preprocessed_test
```

```python
target = "Counterfeit_Sales"
```

```python
x_train = preprocessed_train.drop(target,axis = 1)
y_train = preprocessed_train[target]
```

```python
import xgboost as xgb 
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgb_params = {  
                "learning_rate":[0.01,0.05,0.1,0.3,0.5],
                "gamma":[i/10.0 for i in range(0,5)],
                "max_depth": [2,3,4,5,6,7,8],
                "min_child_weight":[1,2,5,10],
                "max_delta_step":[0,1,2,5,10],
                "subsample":[i/10.0 for i in range(5,10)],
                "colsample_bytree":[i/10.0 for i in range(5,10)],
                "colsample_bylevel":[i/10.0 for i in range(5,10)],
                "reg_lambda":[1e-5, 1e-2, 0.1, 1, 100], 
                "reg_alpha":[1e-5, 1e-2, 0.1, 1, 100],
                "scale_pos_weight":[1,2,3,4,5,6,7,8,9],
                "n_estimators":[100,500,700,1000]
             }
```

```python
Xgb = XGBRegressor(objective = "reg:squarederror")
```

```python
random_search = RandomizedSearchCV(Xgb, n_jobs = -1, cv = 10, n_iter = 50, scoring = 'neg_mean_absolute_error',
                                   param_distributions = xgb_params, verbose = 3)
```

```python
random_search.fit(x_train, y_train)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
report(random_search.cv_results_,5)
```

```python
submissions = pd.DataFrame({target:random_search.predict(preprocessed_test)})
```

```python
submissions.to_csv("my_submision_XGBRegressor.csv", index = False)
```
